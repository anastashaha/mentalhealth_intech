```
#filter out extreme values 
df_without_NaN = df.copy().dropna(subset=["Age"])
df_without_NaN
mean_age = df_without_NaN.mean()["Age"]
mean_age
```

```
df.loc[df["Age"] < 13, "Age"] = mean_age #filling wrong values with mean age
df.loc[df["Age"] > 72, "Age"] = mean_age #filling wrong values with mean age
```

```
def country_encoding(country):
    if country == "United States of America":
        return 1
    elif country == "United Kingdom":
        return 2
    elif country == "Canada":
        return 3
    elif country == "Germany":
        return 4
    elif country == "Netherlands":
        return 5
    elif country == "Australia":
        return 6
    else:
        return 7
```

```
df["Country"] = df["Country"].apply(country_encoding)
```

```
df_curr = df[df["Current Mental Disorder"]=="Yes"]
df_curr
```

# Visualisations

```
#plotting mental disorders based on anonymity  by age
fig, axes = plt.subplots(nrows=1, ncols=2,figsize=(15, 7))
anon = df_curr[df_curr["Anonymity"]=="Yes"]
nonanon = df_curr[df_curr["Anonymity"]=="No"]

#graph US mental disorder
ax = sns.distplot(anon[anon["Sought Treatment"]==1]["Age"].dropna(), bins=20, label = "Sought Treatment",  ax=axes[0], kde=False, color="orange")
ax = sns.distplot(anon[anon["Sought Treatment"]==0]["Age"].dropna(), bins=20, label = "Did Not Seek Treatment", ax=axes[0], kde=False, color="red")
ax.set(ylim=(0, 100))
ax.legend()
ax.set_title("Anon")

#graph non-US mental disorder
ax = sns.distplot(nonanon[nonanon["Sought Treatment"]==1]["Age"].dropna(), bins=20, label = "Sought Treatment",  ax=axes[1], kde=False, color="orange")
ax = sns.distplot(nonanon[nonanon["Sought Treatment"]==0]["Age"].dropna(), bins=20, label = "Did Not Seek Treatment", ax=axes[1], kde=False, color="red")
ax.set(ylim=(0, 40))
ax.legend()
ax.set_title("No Anon")
```

```
#plotting seeking of treatment of those who have family history of MH vs those who don't by age
fig, axes = plt.subplots(nrows=1, ncols=2,figsize=(15, 7))
FH = df[df["MH Family History"]=="Yes"]
no_FH = df[df["MH Family History"]=="No"]

#graph family history treatment
ax = sns.distplot(FH[FH["Current Mental Disorder"]=="Yes"]["Age"].dropna(), bins=3, label = "Has Current Mental Disorder",  ax=axes[0], kde=False, color="orange")
ax = sns.distplot(FH[FH["Current Mental Disorder"]=="No"]["Age"].dropna(), bins=3, label = "Does Not Have Current Mental Disorder", ax=axes[0], kde=False, color="red")
ax.set(ylim=(0, 400))
ax.set(xlim=(0,70))
ax.legend()
ax.set_title("Family History of Mental Health Disorder(s)")

#graph those with no family history treatment
ax = sns.distplot(no_FH[no_FH["Current Mental Disorder"]=="Yes"]["Age"].dropna(), bins=3, label = "Has Current Mental Disorder",  ax=axes[1], kde=False, color="orange")
ax = sns.distplot(no_FH[no_FH["Current Mental Disorder"]=="No"]["Age"].dropna(), bins=3, label = "Does Not Have Current Mental Disorder", ax=axes[1], kde=False, color="red")
ax.set(ylim=(0, 300))
ax.set(xlim=(0,70))
ax.legend()
ax.set_title("No Family History of Mental Health Disorder(s)")
```
